WE will use a tool called "learnPopGen" which was designed to teach you about population genetics with interactive tasks and graphical output.

The first requirement is to get a copy of the R "package" installed locally. The next box is running the R language. TO "run" the box, hit the arrow to the left. That will take a few minutes (<10, but not zero). So, while you're "running" the next box, please scroll down & read about a couple of the tools.

```r
install.packages("learnPopGen")
library(learnPopGen)
```

There is a long list of simulation tools in LearnPopGen- feel free to explore them. But, we will use ***mainly*** selection, coalescence.plot(), and mutation.selection (or msd())

[Just FYI: in R, you tell a series of numbers with c(1,2,3).]

We will start by exploring the effect of SELECTION on allele frequency A1. By default, the starting allele frequency, p0=0.01 or 1% of the population. *You can change that value in the box below.*
The fitness numbers (w) are for the A1A1, A1A2, and A2A2 genotypes. ALWAYS in that order. The largest number for fitness is always 1. Thus:
*   c(1, 1, 0.8) is for a dominant selection in favor of A1A1
*   c(1, 0.8, 0.8) would show recessive for A1A1
*   c(1, 0.8, 0.6) is incomplete dominance for the A1 allele
*   c(0.8, 1, 0.8) is for overdominant inheritance and
*   c(1, 0.8, 1) represents underdominance.

You can change the values for the fitness in the box below & rerun as many times as you want. Note that sometimes a time = 500 (in generations) isn't long enough or is too long. Fell free to adjust the number of generations.

If you explore underdominance or overdominance, you will want to adjust the starting allele frequency for A1 to various values besides 0.1





```r
selection(p0=0.2, w=c(1.0,0.8,1),time=100)
selection(p0=0.3, w=c(1.0,0.7,1),time=100, add=TRUE, color=("red"))
```
